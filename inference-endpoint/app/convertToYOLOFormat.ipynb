```python
import os
from PIL import Image

def convert_to_yolo_format(x, y, w, h, img_width, img_height):
    x_center = (x + w/2) / img_width
    y_center = (y + h/2) / img_height
    w = w / img_width
    h = h / img_height
    return (x_center, y_center, w, h)

dataset_folder =  os.path.join(ROOT_DIR, 'testing')
#'/content/gdrive/My Drive/Process_AMR_Annotations/images'  # Folder containing your dataset images and annotations
output_folder =  os.path.join(ROOT_DIR, 'data/labels/test')
#'/content/gdrive/My Drive/Process_AMR_Annotations/output'  # Folder to save YOLO format annotations

for filename in os.listdir(dataset_folder):
    if filename.endswith(".jpg") or filename.endswith(".jpeg"):
        image_path = os.path.join(dataset_folder, filename)
        annotation_file = os.path.splitext(filename)[0] + '.txt'
        annotation_path = os.path.join(dataset_folder, annotation_file)

        if os.path.exists(annotation_path):
            with Image.open(image_path) as img:
                img_width, img_height = img.size

            with open(annotation_path, 'r') as f:
                for line in f:
                    if line.startswith('position:'):
                        x, y, w, h = map(int, line.strip().split()[1:])
                        x_center, y_center, w, h = convert_to_yolo_format(x, y, w, h, img_width, img_height)

                        # Save to separate .txt files in the output folder
                        with open(os.path.join(output_folder, annotation_file), 'w') as out_f:
                            out_f.write(f"0 {x_center:.4f} {y_center:.4f} {w:.4f} {h:.4f}\n")

print("Conversion complete!")
```
